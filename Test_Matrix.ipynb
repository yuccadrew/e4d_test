## 2. Mesh Generation Requirements (Mode 1/21/31/41)
<a href="#section1">(Previous Section)</a> $\;$ <a href="#section3">(Next Section)</a> $\;$ <a href="#top">(Back to Top)</a>

If _run_mode_ (see _e4d.inp_) is specified with an integer value of 1, E4D will enter mesh generation mode and construct the computational mesh. The requirements for mesh generation mode are specified in Requirement 2.1 through 2.5 below.

#### Requirement 2.0

E4D requires tetgen 1.5.0 installed for the mesh genreation mode. We have noticed big differences in some cases when using tetgen 1.5.0 and tetgen 1.5.1. These differences may cause inaccurate results in the following E4D forward and inversion runs. Tetgen 1.5.0 is included in the git repository for installation.

Subroutine _valid_tetgen_ compared the mesh produced by the running tetgen with the standard mesh produced by tetgen 1.5.0
- 2.0.0 E4D was ran in mode 1 using the _200EW_T.cfg_ input in _e4d.inp_
- 2.0.1 A _check.txt_ was generated to list the different between the element files of the two meshes
- 2.0.2 The _check.txt_ was empty indicating the two meshes were identical

#### Requirement 2.1

In the mesh generation mode, E4D will construct _mesh_build.log_, triangle _*.poly_, and tetgen _*.poly_ files. The tetgen _*.poly_ will have the same prefix as the mesh configuration file. For example, if the mesh configuration file is _my_mesh.cfg_, then the tetgen input file created by E4D will be _my_mesh.poly_, and tetgen will produce mesh files labeled _my_mesh.1.node_ (node file), _my_mesh.1.ele_ (element file), _my_mesh.1.face_ (boundary faces file), and _my_mesh.1.neigh_ (neighbor elements file). Format specifications for each of the meshfiles produced by tetgen can can be found at the tetgen website.

Subroutine _valid_mshout_ checked the existence of the output files for _run_mode_ 1
- 2.1.0 E4D was run in _mode_ 1 using the _two_blocks.cfg_ input in _e4d.inp_
- 2.1.1 A _mesh_build.log_ file was output in the directory
- 2.1.2 A _surface.poly_ file for triangle was output in the directory
- 2.1.3 A _two_blocks.poly_ file for tetgen was output in the directory
- 2.1.4 The mesh files _two_block.1.node_, _two_blocks.1.ele_, _two_blocks.1.neigh_, _two_blocks.1.edge_, and _two_blocks.1.face_ generated by tetgen were output in the directory
- 2.1.5 Visualization files were output as specified in _two_blocks.cfg_. For bx, a _two_blocks.exo_ file generated; for px, _two_blocks.xmf_ and _two_blocks.h5_ were output.

#### Requirement 2.2

If _run_mode_ is 1 (see _e4d.inp_), E4D will open and read the mesh configuration file specified on the _mesh_file_ line of _e4d.inp_. The format of the mesh configuration in file is described in the mesh generation section of the [user guide](https://e4d.pnnl.gov/Documents/E4D_User_Guide.pdf). Upon reading the mesh configuration file, E4D will print read progression messages to the file _mesh_build.log_. The purpose of _mesh_build.log_ is to provide users with information concerning possible formatting errors in the mesh configuration file.

Subroutine _valid_mshlog_ checked the file _mesh_build.log_ for _run_mode_ 1
- 2.2.0 A _mesh_build.log_ file was found in the directory

- 2.2.4 Zone configuration in _mesh_build.log_ and _two_blocks.cfg_ matched

#### Requirement 2.3

E4D will construct a _*.poly_ input file for [triangle](http://www.cs.cmu.edu/~quake/triangle.html), and call triangle in order to construct a high quality 2D triangular mesh, which will be used to construct the surface of the 3D tetrahedral mesh used by E4D. Triangle will be called to produce a mesh with angles no smaller than 30 degrees in each triangle. One mesh node will be placed at each control point specified with a surface flag of 1 in the mesh configuration file. Outer boundaries of the surface will adhere to those specified in the mesh configuration file.

Subroutine _valid_poly2d_ checked the triangle _*.poly_ file for _run_mode_ 1
- 2.3.0 A _surface.poly_ file was found in the directory
- 2.3.1 Surface control points with a boundary flag of 1 in _surface.poly_ and _two_blocks.cfg_ matched
- 2.3.2 Surface control points with a boundary flag of 2 in _surface.poly_ and _two_blocks.cfg_ matched

#### Requirement 2.4

E4D will construct the _*.poly_ input file for [tetgen](http://wias-berlin.de/software/tetgen/) using the surface mesh generated by triangle and the mesh configuration specifications specified in the mesh configuration file. E4D will then call tetgen to build the computational mesh.
Outer boundaries of the mesh will extend vertically downward from the boundaries of the surface mesh. The lower boundary will be horizontal at the depth specified in the mesh configuration file.
The tetgen _*.poly_ file will be constructed to enable creation of inner mesh zones and holes as specified in the mesh configuration file. Elements in each zone will be constructed with a maximum volume constraint as specified in the mesh configuration file. Each distinct zone will be assigned a non-negative integer value.

In order to assign boundary conditions, the following node boundaries will be assigned in the tetgen _*.poly_ file: (1) surface nodes will be assigned a boundary flag of 1, (2) nodes on the outer boundaries of the mesh other than the surface boundary (i.e. 'buried' boundaries) will be assigned a boundary flag of 2, (3) nodes on inner boundaries defining element zones will be assigned a non-negative integer value other than 1 or 2, and (4) nodes on inner boundaries defining electrically conductive infrastructure (assumed infinite conductivity) will be assigned a negative integer value.

## 3. Forward Simulation Requirements (Mode 2/22/32/42)
<a href="#section2">(Previous Section)</a> $\;$ <a href="#section4">(Next Section)</a> $\;$ <a href="#top">(Back to Top)</a>

In E4D, forward simulation refers to the numerical process whereby subsurface electrical potential is simulated given 1) an electrical current transmitted between current source and sink electrodes,  2) a discretized electrical conductivity distribution, and 3) a surface boundary with specified topography. The potential is simulated by solving the Poisson equation as described in the [theory guide](./2010_parallel_ERT.pdf). To produce simulated measurements, the potential simulated at the negative measurement electrode is subtracted from the potential simulated at the positive measurement electrode.  Forward simulation requirements for E4D given in Requirment 3.1 through Requirement 3.6 below.

#### Requirement 3.1

E4D will provide (1) the option to compute the analytic solution for a pole or dipole current source using either surface or buried electrodes within a homogeneous half-space of real electrical conductivity, (2) the option to simulate and output the measurements specified in _survey_file_, and (3) the option to output any, all, or a specified subset of forward simulation results (i.e. simulated potentials) on the computational mesh.

Subroutine _valid_anout_ checked the existence of the output files for _run_mode_ 'analytic'
- 3.1.0 E4D was ran in _mode_ 'analytic' using the survey configuration file _two_blocks.srv_ for a homogeneous half-space model
- 3.1.1 A _two_blocks.sig.srv_ file containing simulated measurements and standard deviations was output in the directory (same format as _two_blocks.srv_)
- 3.1.2 A _two_blocks.dpd_ file containing simulated measurements was output in the directory, as specified in output options file _two_blocks.out_
- 3.1.3 Four simulated potential files were output in the directory, as specified in the output options file _two_blocks.out_

#### Requirement 3.2

E4D simulated analytic solutions will be bechmarked with theoretical values.

Subroutine _valid_ansol_ checked E4D analytic solutions against the theoretical values for _run_mode_ 'analytic'
- 3.2.0 The files listed in 3.1.1 through 3.1.3 were found in the directory
- 3.2.1 Simulated measuremenets in _two_blocks.sig.srv_ and theoretical values matched
- 3.2.2 Simulated potentials in _an_potential.*_ and theoretical values matched

#### Requirement 3.3

E4D will provide the option numerically solve the 3D Poisson equation for a pole or dipole current source with electrodes of arbitrary dimension, and with an arbitrary discretized real conductivity distribution. The solution will be implemented with zero-flux boundary conditions on the surface, and zero-potential boundary conditions on the buried boundaries.

#### Requirement 3.5

E4D will enable numerical forward solutions to be computed in parallel on shared or distributed memory multi-processor computing systems using the Message Passage Interface (MPI) protocol. One processor will be designated the master process, and the remaining processors will be designated as slave processes. The master process will read input and write output, send information to slaves including commands, and receiving information from slave to generate required output. Slave processes will execute the bulk of the computational burden, including executing the numerical forward solutions. Each slave will be assigned one or more current source electrodes, and will compute and store the forward solution for assigned pole. Full field dipole solutions will be assembled by the master node after requesting and receiving the pole solutions from the appropriate slaves. Simulated survey measurements will also be assembled by the master after receiving the simulated electrode potentials from the appropriate. Pole solution computations will be distributed evenly among slaves. Thus, the maximum number of processors that can be used for a given ERT survey simulation is _ne+1_, where _ne_ is the number of electrodes.
